## pandas series

```python
# series from list 
country=["india","pakistan","Usa"]
a=pd.Series(country)
type(a)
a

```

```python
# custome index 
marks = [ 99,99,99,99]
subject=["Dsa","Data science","OOPS","Machine Learning"]
pd.Series(marks,index=subject,name="Aditya ke marks")
```
