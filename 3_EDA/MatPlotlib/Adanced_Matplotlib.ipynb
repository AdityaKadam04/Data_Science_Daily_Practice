# Advanced Matplotlib


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## Scatter Plot Advanced colored

```python
df = pd.read_csv("iris.csv")
df.head(5)
```

```python
plt.scatter(df["SepalLengthCm"],df["PetalLengthCm"])
plt.title("Sepal lenght vs petal lenght")
plt.xlabel("sepal lenght")
plt.ylabel("petal lenght")
```

```python
df["Species"].sample(5)
```

```python
# Assign number to species 
plt.figure(figsize=(12,7))
df["Species_encode"]=df["Species"].replace({'Iris-setosa':0 ,'Iris-versicolor':1,"Iris-virginica":2})
plt.scatter(df["SepalLengthCm"],df["PetalLengthCm"],c=df["Species_encode"],cmap="summer",alpha=1)
plt.title("Sepal lenght vs petal lenght")
plt.xlabel("sepal lenght")
plt.ylabel("petal lenght")
plt.show()
```

```python
# own dataset
df=pd.read_csv("batter.csv")
s_df=df.head(100).sample(25,random_state=5)
plt.scatter(s_df["avg"],s_df["strike_rate"])
for i in range(s_df[0]):
    plt.text(s_df["avg"].values[i],s_df["strike_rate"].values[i],df["batter"].values[i])
```
