# MatPlotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("classic")
```

## 2D Line Plot

```python
# Bivariate Analysis , Num-Num , cat-cat, Num , Num-cat , use case : -Time series Data 
# Plotting a simple Function
price=[48,54,57,49,15,10]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
# From Pandas DataFrame 
df = pd.read_csv("sharma-kohli.csv")
df
```

```python
plt.plot(df["index"],df["V Kohli"],color="Black",linestyle = "solid",linewidth=3,marker="D",label="Virat")
# Multiple Plotting
plt.plot(df["index"],df["RG Sharma"],color = "blue",linestyle = "solid",linewidth=3,marker="D",label="Rohit")

# label title 
plt.title("Rohit Sharma vs Virat Kohli Comparison")
plt.xlabel("Season")
plt.ylabel("Run score ")

plt.legend()
plt.grid()
plt.show()
```

```python
# Limiting axes : Agar data kissi Axes pe bohot bada ho jaye to hum axes ko limit laga dete hai 
price=[48,120,60,5,87,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.ylim(0,400)
plt.xlim(2014,2020)
plt.show()
```

## ScatterPlot

```python
# Bivariate Analysis
# Numerical vs Numerical 
# use case: - correlation between 2 variable 
# Plot Scatter Plot using simple function
x=np.linspace(-10,10,10)
y=10*x+3 +np.random.randint(0,300,10)
plt.scatter(x,y)
```

## Plot Scatter On Pandas Data 

```python
df = pd.read_csv("batter.csv")
df=df.head(50)
plt.scatter(df["avg"],df["strike_rate"],color = "Red")
plt.title("Avg and sr analytics")
plt.xlabel("Average")
plt.ylabel("strike rate")
plt.grid()
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
plt.scatter(tips["total_bill"],tips["tip"],s = tips["size"]*20)
# use s for kitne log sath aaye hoge 
plt.title("Correlation between Total bill and tip")
plt.xlabel("Total_bill")
plt.ylabel("tips")
plt.grid()
plt.show()
```

# Multiple Bar Chart 

```python
batsman=pd.read_csv("batsman_season_record.csv")
batsman.head(5)
```

```python
plt.bar(np.arange(batsman.shape[0]),batsman["2016"])
```

```python
# Problems With Bar Chart  : Name Overlapping problem
batsman.head(2)
```

```python
plt.bar(batsman["batsman"],batsman["2017"],label ="2017")
plt.bar(batsman["batsman"],batsman["2016"],bottom=batsman["2017"],label ="2016")
plt.bar(batsman["batsman"],batsman["2015"],bottom=batsman["2017"]+batsman["2016"],label="2015")
plt.legend()
plt.show()
```

## Pie Chart 

```python
# Univariate Vs Bivariate analysis  , categorical vs numerical , use case : to find contribution on standard scale of 100
# simple data 
data = [23,45,100,20,49]
marks = ["eng","math","sci","ss","history"]
plt.pie(data,labels=marks)
plt.legend()
```
